## Алгоритм по схожести навыков

```python
from collections import Counter
from itertools import combinations


pair_freq = Counter()

for lst in hh_skills:
    # Генерируем все возможные пары слов в списке
    pairs = combinations(sorted(lst), 2)
    # Обновляем счетчик пар
    pair_freq.update(pairs)

pair_freq
```

```python
pair_freq_df = pd.DataFrame(pair_freq.items(), columns=['pair', 'frequency']) # создаём dataframe частоты
```

```python
pair_freq_df
```

## Создание функции для подбора тегов

```python
def convert_tuple_to_str(value: tuple, tag: str):
    new_tag = list(filter(lambda x: x != tag, value))[0]
    
    return new_tag


def get_similar_tags(user_tags: list, tags_count: int):
    tags_list = []
    for user_tag in user_tags:
        count = int(tags_count/len(user_tags))
        
        similar_tags = pair_freq_df[pair_freq_df["pair"].apply(
            lambda x: (user_tag in x) and not (x[0] in user_tags and x[1] in user_tags))].sort_values(by="frequency", ascending=False)

        for similar_tag in similar_tags["pair"]:
            tag = convert_tuple_to_str(similar_tag, user_tag)

            if tag not in tags_list:
                tags_list.append(tag)
                count -=1

            if count == 0:
                break
                
    
    return tags_list
```

## Попытка улучшить модель

### Создание матрицы совместного употребления

```python
from collections import defaultdict


occurrence_dict = defaultdict(lambda: defaultdict(int)) # Создание 2-уровнего словаря

for lst in hh_skills:
    pairs = combinations(lst, 2) # создание всевозможных пар для каждой строки со скиллами
    
    for tag1, tag2 in pairs:
        if tag1 in user_skills and tag2 in user_skills:
            occurrence_dict[tag1][tag2] +=1
            occurrence_dict[tag2][tag1] +=1

occurence_df = pd.DataFrame(occurrence_dict).fillna(0)

occurence_df
```

### Создание матрицы косинусного сходства навыков

```python
from sklearn.metrics.pairwise import cosine_similarity


cosine_sim_matrix = pd.DataFrame(cosine_similarity(occurence_df.T), 
                                 index=occurence_df.columns, 
                                 columns=occurence_df.columns)

print("Cosine Similarity Matrix:")
print(cosine_sim_matrix)
```

```python
print(cosine_sim_matrix["Docker"].sort_values(ascending=False))
```

### Создание функции среднего косинусного сходства между выбранными тегами и новым

```python
def get_average_cosine_similarity(tag, tag_list, cosine_sim_matrix):
    similarities = []
    
    for user_tag in tag_list:
        if user_tag in cosine_sim_matrix.index and user_tag in cosine_sim_matrix.columns:
            similarities.append(cosine_sim_matrix.loc[tag, user_tag])

    return np.mean(similarities) if similarities else 0
```

### Создание функции по подбору похожих навыков

```python
def get_similar_tags(tag_list, tag_count):
    tag_df = pd.DataFrame(cosine_sim_matrix.index, columns=["Name"])

    tag_df["Score"] = tag_df["Name"].apply(get_average_cosine_similarity, args=(tag_list, cosine_sim_matrix))

    result_tags = tag_df[~tag_df["Name"].isin(tag_list)].sort_values(by="Score", ascending=False)[:tag_count]

    return list(result_tags["Name"].values)
```
